## Вопрос 1.

Да, алгоритмы могут обучаться, и это процесс, при котором модель на основе данных настраивает свои параметры для выполнения конкретной задачи. Процесс обучения включает в себя два основных типа: обучение с учителем (supervised learning), где модель обучается на размеченных данных, и обучение без учителя (unsupervised learning), где модель работает с неразмеченными данными.

## Вопрос 2.

Модель линейной регрессии — это статистический метод, который используется для моделирования линейной зависимости между одной или несколькими независимыми переменными (предикторами) и зависимой переменной (ответом). Основная цель — найти линию, которая наилучшим образом аппроксимирует данные и позволяет предсказывать значения ответа на основе новых значений предикторов.

## Вопрос 3.

В градиентном спуске параметры модели подбираются путем последовательного обновления значений, чтобы минимизировать функцию потерь. На каждом шаге вычисляется градиент функции потерь по параметрам модели, и они обновляются в направлении, противоположном градиенту, что позволяет постепенно уменьшать ошибку модели. Процесс повторяется до достижения минимума функции потерь.

## Вопрос 4.

Темп обучения (learning rate) — это величина шага, с которой обновляются параметры модели в процессе градиентного спуска. Если темп слишком мал, обучение будет происходить медленно и может потребовать много итераций. Если же темп слишком велик, модель может перескакивать через минимум и не сойтись к оптимальному решению или даже расходиться. Оптимальный темп обучения зависит от конкретной задачи и может быть настроен экспериментально.
